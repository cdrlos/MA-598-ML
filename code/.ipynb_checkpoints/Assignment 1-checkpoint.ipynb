We'll start with the simple convnet used in section 5.1 as a baseline.  It's a stack of `Conv2D` and `MaxPooling2D` layers. As we've seen, a convnet takes as input tensors of shape `(image_height, image_width, image_channels)` (not including the batch dimension). 
In our case, we will configure our convnet to process inputs of size `(32, 32, 3)`, which is the format of CIFAR10 images (32x32 images with 1 channel for each of red, green, blue). We do this by 
passing the argument `input_shape=(32, 32, 3)` to our first layer.

As before, we then flatten the output and feed it to a fully connected layer with softmax output to produce a probability distribution over the 10 output classes.

```python
from keras import layers
from keras import models

layer_size = [32, 64, 64, 64]

model = models.Sequential()
model.add(layers.Conv2D(layer_size[0], (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(layer_size[1], (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(layer_size[2], (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(layer_size[3], activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
model.summary()
```

In this case, the `(4, 4, 64)` outputs were flattened into vectors of length `1024` before going through two `Dense` layers.

Next, we'll load the CIFAR10 images. We will reuse a lot of the code we have already covered in the MNIST example from Chapter 
2.

```python
from keras.datasets import cifar10
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()

train_images = train_images.reshape((50000, 32, 32, 3))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 32, 32, 3))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

We'll set up some plotting functions to help visualize the results.  

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot multiple images in rows
def plot_activation(output):
  num_cols = 8
  num_rows = np.int(np.ceil(output.shape[-1]/num_cols))
  fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(9.3, 6),
                        subplot_kw={'xticks': [], 'yticks': []})

  fig.subplots_adjust(left=0.03, right=0.97, hspace=0.3, wspace=0.05)

  for ax, ind in zip(axs.flat, np.arange(output.shape[-1])):
    ax.imshow(np.squeeze(output[0,:,:,ind]))
    ax.set_title(str(ind))

  plt.tight_layout()
  plt.show()  

# Plot a single image  
def plot_image(image_array, size=1.5):
  plt.imshow(np.squeeze(image_array), interpolation='none', aspect='equal')
  ax = plt.gca()
  ax.grid(b=None)
  fig = plt.gcf()
  fig.set_size_inches(size, size)
  ax.axes.xaxis.set_ticklabels([])
  ax.axes.yaxis.set_ticklabels([])
  return fig, ax

  
```

Now we can compile the model.   For a reality check, we'll print out the shape of one of the layers of weights.  The first layer is layer 0 - the kernels are 3x3 and include all three RGB channels, which gives 3x3x3.  There are 32 of these filters, so we get 3x3x3x32.

```python
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

cur_layer = model.layers[0]
orig_wts = cur_layer.get_weights()
wts0 = np.asarray(orig_wts[0])
print(wts0.shape)

```

Now we'll test the model before training.  With 10 classes, we expect an accuracy of about 10%.  

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
test_acc
```

We next set up a model that provides intermediate output.  This allows us to display the activations for a given layer.  You can change the input image index and which layer to display.  

```python
from keras.models import Model

layer_index = 0  # Change this to change which layer of activations are visualized
image_ind = 11  # Change this to change which image is visualized.  

intermediate_layer_model = Model(inputs=model.input,
                                 outputs=model.layers[layer_index].output)


intermediate_output = intermediate_layer_model.predict(test_images[image_ind:image_ind+1, :, :, :])
print(intermediate_output.shape)

plot_image(test_images[image_ind])
plot_activation(intermediate_output)

# Note that you can modify the code above to plot weights instead of activations.  
```

Now we can train the model.  We'll use validation_split to leave out some of the training data for evaluation during training.  This is in addition to the test data that we'll use for final evaluation.   

```python

history = model.fit(train_images, train_labels, validation_split=0.1, epochs=10, batch_size=64)
```

Plot the loss and accuracy as a function of epoch. 

```python
# list all data in history
print(history.history.keys())

def plot_training(history):
  # summarize history for loss
  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('model loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()
  # summarize history for accuracy
  plt.plot(history.history['acc'])
  plt.plot(history.history['val_acc'])
  plt.title('model accuracy')
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(['train', 'test'], loc='upper left')
  plt.show()

plot_training(history)
```

Let's evaluate the trained model on the test data:

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Accuracy on test images is ' + str(test_acc))
```

After 10 epochs, the accuracy increased from about 10% to 60% or more.  We can view the activations after training. We can also show the classification for a few images.  

```python
num_images = 24  # number of images to classify and display
image_ind = 150  # index of starting image

# Convert to format needed for prediction
k = np.array(test_images[image_ind:image_ind+num_images])
y = k.reshape(num_images,32,32,3)

# Get the predictions and get top predictions
prediction = model.predict(y)
top_inds = prediction.argsort()
top_inds = np.flip(top_inds, axis=1)  # argsort sorts in ascending order, so flip to get descending

names = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# The code below is similar to the activation plotting function above, and could 
# be refactored to avoid duplication.  However, here we do more processing to 
# determine the text associated with each image.  
num_cols = 6
num_rows = np.int(np.ceil(num_images/num_cols))
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(10.3, 3*num_rows),
                      subplot_kw={'xticks': [], 'yticks': []})

# For each image, determine the correct label and the top 5 predictions.  
# Use these in text below the image.  
num_correct = 0
for ax, ind in zip(axs.flat, np.arange(num_images)):
  cur_ind = image_ind+ind
  
  correct_label_ind = test_labels[cur_ind].argsort()[-1]
  pred_text = names[correct_label_ind] + '\n'
  
  top_label_ind = top_inds[ind, 0]
  if top_label_ind == correct_label_ind:
    num_correct = num_correct + 1
    
  for pred_ind in top_inds[ind, 0:5]:
    pred_text = pred_text + names[pred_ind] + ', prob = ' + str(round(prediction[ind, pred_ind],2)) + '\n'
  ax.imshow(np.squeeze(test_images[cur_ind]))
  ax.set_xlabel(pred_text)

fig.subplots_adjust(left=0.03, right=0.97, hspace=0.3, wspace=0.05)

plt.tight_layout()
plt.show()  

print('Fraction correct = ' + str(num_correct) + '/' + str(num_images))
print('Percent correct = ' + str(round(100*num_correct/num_images, 1)))
  
```
